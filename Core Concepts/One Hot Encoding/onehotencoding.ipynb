# 2. k-1 OneHotEncoding

##### K-1 encoding is used to convert a categorical variable with K categories into K-1 dummy variables. It is mainly used in Machine Learning (especially Linear Regression) to avoid the Dummy Variable Trap (multicolarity problem).

```python
pd.get_dummies(df, columns=['fuel', 'owner'], drop_first=True)
```
