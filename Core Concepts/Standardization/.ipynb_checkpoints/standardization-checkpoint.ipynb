```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns
```

```python
df= pd.read_csv("Social_Network_Ads.csv")
```

```python
df.sample(2)
```

```python
df= df.iloc[:, 2:]
```

```python
df.sample(2)
```

```python
X= df.iloc[:, 0:2]
y= df.iloc[:, -1]
```

  # Train Test Split
 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.1, random_state=3)
X_train.shape, X_test.shape
```

# StandardScaler

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(X_train) # StandardScaler only cares about the input features X_train, not the target y_train.

X_train_scaled= scaler.transform(X_train)
X_test_scaled= scaler.transform(X_test)
```

```python
X_train
```

```python
X_train_scaled
```

##### X_train is a pandas df and X_train_scaled is numpy array so we convert that array into df. Same for X_test

```python
X_train_scaled= pd.DataFrame(X_train_scaled, columns= X_train.columns)
X_test_scaled= pd.DataFrame(X_test_scaled, columns= X_test.columns)
```

```python
np.round(X_train.describe(), 2)
```

##### Standardization is a feature scaling method that scales data such that mean=0 and std=1

```python
np.round(X_train_scaled.describe(), 2)
```

# Effect of Scaling

```python
fig, (ax1, ax2)= plt.subplots(ncols=2, figsize=(12,4))
ax1.scatter(X_train['Age'], X_train['EstimatedSalary'], color='green')
ax1.set_title("Before Scaling")
ax2.scatter(X_train_scaled['Age'], X_train_scaled['EstimatedSalary'], color='red')
ax2.set_title("After Scaling")
sns.despine()
plt.show()
```

```python
fig, (ax1, ax2)= plt.subplots(ncols=2, figsize=(12,6))
sns.despine()
ax1.set_title("Before Scaling")
sns.kdeplot(data= X_train, ax= ax1, fill=True)
ax2.set_title("After Scaling")
sns.kdeplot(data=X_train_scaled, ax=ax2, fill=True)
```
