# Extracting Numerical Part

```python
df['number_numerical']= pd.to_numeric(df["number"], errors= 'coerce', downcast='integer')
# It converts the number column into numeric form and stores it in a new column number_numerical.
# errors='coerce' If a value cannot be converted, it becomes NaN instead of throwing an error.
```

# Extracting Categorical Part

```python
df['number_categorical']= df['number'].where(df['number_numerical'].isna())
```

```python
df['cabin_num']= df['Cabin'].str.extract('(\d+)')  # extracts numerical part.
df['cabin_cat']= df['Cabin'].str[0] # extract first letter as cabin name has first letter in categorical form.
```
