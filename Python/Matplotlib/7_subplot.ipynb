```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array([1,2,3,4])
y= np.array([10,20,15,25])

plt.subplot(1,2,1) #1st row, 2nd column, 1st subplot
plt.bar(x,y)
plt.title("Bar chart")

plt.subplot(1,2,2) #1st row, 2nd column, 2nd subplot
plt.plot(x,y)
plt.title("Line chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array([1,2,3,4])
y= np.array([10,20,15,25])

#fig, ax= plt.subplots(nrows, ncols, figsize=(width, height))
fig, ax= plt.subplots(1,2, figsize=(10,5), sharex=True, sharey= True)    # shares x axis you can see in figure x axis value is same
ax[0].plot(x,y, color="red")
ax[0].set_title("Line Plot")

ax[1].bar(x,y, color="green")
ax[1].set_title("Bar Chart")

plt.tight_layout()
plt.show()
```
