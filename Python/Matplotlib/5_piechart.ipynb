```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array(["Apples", "Bananas", "Mangoes", "Oranges", "Kiwis"])
y= np.array([2, 1, 4, 3, 2])
plt.pie(y, labels= x, startangle=90, colors=["blue", "green", "pink", "yellow", "orange"], autopct="%1.2f%%") 
# 1.2f → means:
# 1 → minimum width (at least 1 digit before decimal).
# .2f → floating-point number with 2 decimal places.
# The last %% → prints a literal % sign.
plt.legend(loc="upper right")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.array(["Apples", "Bananas", "Mangoes", "Oranges", "Kiwis"])
y= np.array([2, 1, 4, 3, 2])
myexplode= [0.2, 0.1, 0, 0, 0.5]
plt.pie(y, labels= x, startangle=90, colors=["blue", "green", "pink", "yellow", "orange"], autopct="%1.2f%%", explode= myexplode, shadow= True) 
# 1.2f → means:
# 1 → minimum width (at least 1 digit before decimal).
# .2f → floating-point number with 2 decimal places.
# The last %% → prints a literal % sign.
plt.legend(title="Fruits:", loc= "upper left")

plt.show()
```
