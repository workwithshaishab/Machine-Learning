```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

y= np.array([2500,3500,2200,1800,3500,4000,2800])
x = np.array(["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"])

# Different line styles => '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
plt.plot(x,y, color="red", marker="o", linestyle=":", linewidth=3, label="1 Aug - 7 Aug")
plt.title("Sales Data of a Bakery Shop")
plt.xlabel("Week Days")
plt.ylabel("Amount of sales in a day")
plt.legend(loc="upper left", fontsize=10)
plt.grid(color="gray", linestyle=":", linewidth=1)
plt.ylim(0,5000)
plt.xticks(["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"], ["Sun","Mon","Tues","Wed","Thurs","Fri","Sat"])


plt.savefig("bakery-sales-line-plot.png", dpi=300, bbox_inches="tight")  # it should come before plt.show() otherwise, some backends may save a blank figure.
plt.show()
```
