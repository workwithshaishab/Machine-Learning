```python
import matplotlib.pyplot as plt 
import numpy as np

x= np.array(["A", "B", "C", "D"])
y= np.array([1000, 1500,800,1200])

plt.title("Sales of the products")
plt.bar(x, y, color= "orange", label= "Sales 2025", width= 0.5)   # for horizontal bars use plt.barh()
plt.xlabel("Products")
plt.ylabel("No of sales")
plt.legend()
plt.ylim(0,2000)
plt.show()
```
