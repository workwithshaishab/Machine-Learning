```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue="island" )
sns.despine()
```

```python
sns.swarmplot(data=penguins, x="species", y="body_mass_g", hue="island", dodge=True)
sns.despine()
sns.set_context("poster")   #paper, notebook, poster, talk   Default is notebook
```
