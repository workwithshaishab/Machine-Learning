import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")

```python
sns.set_style("whitegrid")
sns.lineplot(data=penguins, x="flipper_length_mm", y="body_mass_g", color="green")
```

```python
sns.set_style("whitegrid")
sns.lineplot(data=penguins, x="flipper_length_mm", y="body_mass_g", color="green", hue="sex")
```

```python
sns.set_style("whitegrid")
sns.lineplot(data=penguins, x="flipper_length_mm", y="body_mass_g", color="green", hue="species")
```
