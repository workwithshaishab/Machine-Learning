```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", color="green")
```

```python
sns.set_style("whitegrid")
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="sex")
```

```python
sns.set_style("whitegrid")
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="sex", kind="kde")    # kernel density
```

```python
sns.set_style("whitegrid")
sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="sex", kind="scatter")   
```

```python

sns.jointplot(data=penguins, x="flipper_length_mm", y="body_mass_g", kind="reg")  # The value for `kind` must be one of ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']
sns.set_style("whitegrid")
```
