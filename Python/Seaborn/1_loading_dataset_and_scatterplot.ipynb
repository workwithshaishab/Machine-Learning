```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="island")
```

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
```

```python
sns.set_style("ticks")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
```

```python
sns.set_style("whitegrid")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
```

```python
sns.set_style("dark")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
```

```python
sns.set_style("darkgrid")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
```

```python
sns.set_style("ticks")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
sns.despine()  #removes top and right side line 
```

```python
sns.set_style("ticks")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="sex")
sns.despine(left=True, bottom= True)  #removes top and right side line   left=True removes left line as well    bottom= True removes bottom line as well
```

```python
sns.set_style("ticks")
sns.set_context("paper")   #paper, notebook, poster, talk   Default is notebook
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species", palette= "Blues") #different palettes are there
sns.despine()  #removes top and right side line 
```

```python
sns.set_style("ticks")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species", alpha=0.5)
sns.despine()  #removes top and right side line 
```
