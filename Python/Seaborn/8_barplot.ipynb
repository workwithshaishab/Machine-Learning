```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette=["pink", "skyblue"])
sns.despine()
```

```python
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette=["pink", "skyblue"], estimator= np.std)  #stdev
sns.despine()
```

```python
sns.barplot(data=penguins, x="species", y="body_mass_g", hue="sex", palette=["pink", "skyblue"], estimator= np.sum)
sns.despine()
sns.set_context("poster")
```
