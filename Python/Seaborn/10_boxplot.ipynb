```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.boxplot(data= penguins, x= "species", y= "body_mass_g")
sns.despine()
```

```python
sns.boxplot(data= penguins, x= "species", y= "body_mass_g", hue="sex", palette=["red", "green"])
sns.despine()
```
