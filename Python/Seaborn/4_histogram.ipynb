```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.set_context("notebook")
sns.histplot(data= penguins, x= "body_mass_g", kde=True) # kde= True draws the curve 
```

```python
sns.set_context("notebook")
sns.histplot(data= penguins, x= "body_mass_g", hue="sex")
sns.despine()
```

```python
sns.set_context("notebook")
sns.histplot(data= penguins, x= "body_mass_g", hue="sex", multiple="stack")
```
