```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.regplot(data=penguins, x="flipper_length_mm", y="body_mass_g", color="green")
sns.despine()
```

```python
sns.regplot(data=penguins, x="flipper_length_mm", y="body_mass_g", scatter=False)
sns.despine()
```
