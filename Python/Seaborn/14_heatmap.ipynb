```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
columns=["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
penguins[columns].corr()
```

```python
columns=["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
sns.heatmap(data= penguins[columns].corr(), annot=True, cmap="Blues")   #annot=true gives value in heatmap
```
