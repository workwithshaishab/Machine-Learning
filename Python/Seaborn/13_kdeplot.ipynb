```python
# Kernel Density Estimate (KDE) — basically a smooth curve that estimates the probability density of your data.

import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.kdeplot(data= penguins, x= "body_mass_g")
sns.despine()
```

```python
sns.kdeplot(data= penguins, x= "body_mass_g", hue="species", fill=True)
sns.despine()
```
