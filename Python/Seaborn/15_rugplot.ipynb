```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.rugplot(data= penguins, x= "body_mass_g", y= "flipper_length_mm", hue="species")
sns.despine()
```

```python
sns.rugplot(data= penguins, x= "body_mass_g", y= "flipper_length_mm", hue="species", height= 0.2)
sns.despine()
```
