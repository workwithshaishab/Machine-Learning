```python
# Each category’s values are shown as points jittered (to avoid overlap).
# Plots one categorical variable vs one numeric variable.

import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.stripplot(data=penguins, x="species", y="body_mass_g", hue="island" )
sns.despine()
```

```python
sns.stripplot(data=penguins, x="species", y="body_mass_g", hue="island", dodge=True)
sns.despine()
```
