```python
import seaborn as sns
import pandas as pd
import numpy as np
sns.get_dataset_names()
penguins= sns.load_dataset("penguins")
```

```python
sns.pairplot(data= penguins)
```

```python
sns.pairplot(data= penguins, hue="sex", diag_kind="kde", height= 4)
```

```python
sns.pairplot(data= penguins, hue="sex", diag_kind="hist", height= 4)
"""
Use "hist" when you want raw distributions.
Use "kde" when you want smooth density curves.
Use "auto" to let Seaborn decide.
Use "none" if you don’t want diagonal plots at all.
"""
```
