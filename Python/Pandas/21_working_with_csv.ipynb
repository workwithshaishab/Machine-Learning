**3. Opening a CSV file from an URL**

```python
import requests
import pandas as pd
from io import StringIO

url = "https://gist.githubusercontent.com/kevin336/acbb2271e66c10a5b73aacf82ca82784/raw/employees.csv"

headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"
}

req = requests.get(url, headers=headers)
req.raise_for_status()  # optional but recommended

data = StringIO(req.text)
df = pd.read_csv(data)

df.head()
```

**7. Skip bad lines**

```python
df= pd.read_csv("bad_lines.csv")
print(df)
```

```python
import pandas as pd
df = pd.read_csv("bad_lines.csv", engine="python", quoting=3, on_bad_lines="skip")       # disable quote handling

# keep only the expected columns
df = df[["id", "name", "age", "city"]]

# drop rows where id or name is missing
df = df.dropna(subset=["id", "name","age", "city"])
print(df)
```

**10. Converters**

```python
def rename(name):
    if name== "Automatic":
        return "Auto"
    else:
        return name
    
df= pd.read_csv("bmw.csv", converters={"transmission":rename})
print(df)
```

**11. Loading a huge dataset in chunks**

```python
df= pd.read_csv("bmw.csv", header=0, chunksize=1200)
for chunks in df:
    print(chunks.shape)
```
