```python
import pandas as pd
from sqlalchemy import create_engine

# 1. Create the engine
# Format: mysql+pymysql://user:password@host/database
engine = create_engine("mysql+pymysql://root:@localhost/world")

# 2. Use the engine connection with Pandas
query = "SELECT * FROM CITY"
df = pd.read_sql_query(query, engine)

# Display the first few rows
print(df.head())
```

```python
pd.read_sql_query("SELECT * FROM city WHERE Population>1000000", engine)
```

```python
pd.read_sql_query("SELECT  Name, GNP, Continent, Region FROM country WHERE Name= 'Nepal'", engine)
```
