```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
```

```python
df= pd.read_csv("customer.csv")
df.sample(5)
```

```python
data= df.iloc[:, 3:5]
data
```

```python
plt.figure(figsize=(10,6))
plt.title("Customer Dendrogram")
shc.dendrogram(shc.linkage(data, method='ward') ) 
# linkage defines how the distance between clusters is calculated when deciding which clusters to merge next.
plt.show()
```

```python
model= AgglomerativeClustering(n_clusters=5, linkage='ward')
pred_data= model.fit_predict(data)
print(pred_data)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(data.iloc[:,0], data.iloc[:,1], c= pred_data, cmap='rainbow') 
plt.show()
```
