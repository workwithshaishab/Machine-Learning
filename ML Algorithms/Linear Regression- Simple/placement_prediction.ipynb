```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("placement.csv")
df.head()
```

# Scatter plot (Understanding the data)

```python
plt.figure(figsize=(12,8))
plt.scatter(df["cgpa"], df["package"])
plt.xlabel("CGPA")
plt.ylabel("Package in LPA")
```

###### This is sort of Linear but not completely linear

```python
X= df.iloc[:, 0:1]
y= df.iloc[:, -1]
```

```python
X
```

```python
y
```

# Train test split


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train, y_train)    # fit helps to train the data
```

```python
X_test
```

```python
y_test
```

# Predicted Values using Linear Regression

```python
y_pred = lr.predict(X_test.iloc[[0]])  
print("Predicted value:", y_pred)
```

# Regression Line

```python
plt.figure(figsize=(12,8))
plt.scatter(df["cgpa"], df["package"])
plt.plot(X_train, lr.predict(X_train), color="green", label= "Regression Line")
plt.legend()  #the label wont work if we dont use this function
plt.xlabel("CGPA")
plt.ylabel("Package in LPA")


# This was the best-fit line learned by the model using Linear Regression
```

```python
m= lr.coef_
b= lr.intercept_
# y= mx+b
m*8.58+b
```
