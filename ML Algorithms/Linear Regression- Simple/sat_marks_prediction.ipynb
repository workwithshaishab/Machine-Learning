```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("sat_marks.csv")
```

```python
df.sample(5)
```

# Scatter plot (Understanding the data)

```python
plt.figure(figsize=(12,8))
plt.scatter(df["GPA"], df["SAT"])
plt.xlabel("GPA")
plt.ylabel("SAT MARKS")
```

```python
X = df.iloc[:,0:1]
X
```

```python
y= df.iloc[:,-1]
y
```

# Train test split


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train, y_train)
```

```python
X_test
```

```python
y_test
```

# Predicted values using Linear Regression

```python
pred= lr.predict(X_test.iloc[[0]])
print("Predicted values:", pred)
```

```python
pred= lr.predict(X_test.iloc[[1]])
print("Predicted values:", pred)
```

```python
pred= lr.predict(X_test.iloc[[2]])
print("Predicted values:", pred)
```

```python
pred= lr.predict(X_test.iloc[[3]])
print("Predicted values:", pred)
```

# Regression Line

```python
plt.figure(figsize=(12,8))
plt.scatter(df["GPA"], df["SAT"])
plt.plot(X_train, lr.predict(X_train), color="red", label="Regression Line")
plt.legend()
plt.xlabel("GPA")
plt.ylabel("SAT MARKS")
```
