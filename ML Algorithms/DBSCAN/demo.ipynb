```python
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import numpy as np
```

```python
X= np.array([[1, 2], [2, 2], [2, 3],[8, 7], [8, 8], [25, 80]])
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
db= DBSCAN(eps= 3, min_samples=2)
db.fit(X)
db.labels_

# noise as -1
```

```python
db= DBSCAN(eps= 3, min_samples=3)
db.fit(X)
db.labels_

# since min sample is 3 so it detected those 2 points as noise
```

```python
db= DBSCAN(eps= 10, min_samples=2)
db.fit(X)
db.labels_

# distance was more so it get grouped in same cluster
```

```python
X, _ =make_circles(n_samples=500, factor=.5, noise=.03, random_state= 42) 
# _→ labels (ignored here)
# If factor = 0.5 → inner circle is half the size
# Smaller factor → tighter inner circle
db= DBSCAN(eps= 0.2, min_samples=5)
clusters= db.fit_predict(X)

plt.scatter(X[:,0], X[:,1], c= clusters)
plt.show()
```
