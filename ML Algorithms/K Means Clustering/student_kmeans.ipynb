```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df= pd.read_csv('student_clustering.csv')
df.sample(3)
```

```python
plt.scatter(x= df['cgpa'], y= df['iq'])
sns.despine()
plt.show()
```

```python
wcss= []
for i in range(1,11):
    model= KMeans(n_clusters= i)
    model.fit_predict(df)
    wcss.append(model.inertia_)  # It gives the WCSS (Within-Cluster Sum of Squares).
print(wcss)
```

```python
plt.plot(range(1,11), wcss)
plt.xlabel("Number of clusters (k)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
sns.despine()
plt.show()
```

# K Means Clustering

```python
X= df.iloc[:,:].values
model= KMeans(n_clusters=4, random_state= 42)
y_pred= model.fit_predict(X)
y_pred
```

# Data in cluster 1

```python
print(X[y_pred== 1])
```

# 1st feature Data in cluster 1

```python
print(X[y_pred == 1, 0])
print(X[y_pred == 1, 1])
```

```python
plt.scatter(X[y_pred== 0,0], X[y_pred == 0,1])
plt.scatter(X[y_pred== 1,0], X[y_pred == 1,1])
plt.scatter(X[y_pred== 2,0], X[y_pred == 2,1])
plt.scatter(X[y_pred== 3,0], X[y_pred == 3,1])
sns.despine()
plt.title("Different Clusters")
plt.show()
```

```python
centers = model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', marker='X', s=200, label='Centroids')
plt.legend()
sns.despine()
plt.title("Visualization of centroids")
plt.show()
```
